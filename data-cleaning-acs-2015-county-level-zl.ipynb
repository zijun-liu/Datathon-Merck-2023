# Clean acs social data 2015

```python
df_social = pd.read_csv(path +"acs_social_5yr_2015_cbsa/ACSDP5Y2015.DP02-Data.csv",
                       skiprows=1)
```

```python
df_social.head()
```

```python
df_social.columns.tolist()
```

```python
# select percent of the following info
# Extract columns where column names contain values in the list
social_lst = ['Married-couple family', 
              '9th to 12th grade, no diploma', 'High school graduate',
              'Some college, no degree', "Associate's degree"] 
```

```python
selected_columns = df_social.columns[df_social.columns.str.contains('|'.join(social_lst))]
cols = [column for column in selected_columns \
        if 'Percent!!' in column and "Annotation" not in column and "18" not in column]
cols
```

```python
df_social = df_social[['Geography','Geographic Area Name'] + cols]
df_social
```

```python
cols = [col.rsplit("!!", 1)[-1] for col in cols]
cols
```

```python
# rename columns
df_social.columns = ['Geography','Geographic Area Name'] + cols

df_social
```

```python
# convert values from percentage to float
for col in cols:
    df_social[col] = df_social[col].astype('float') / 100.0
df_social.head()
```

```python
# rename and bin columns
df_social['marriage_prop'] = df_social['Married-couple family']
df_social['high_school_prop'] = df_social['9th to 12th grade, no diploma'] + df_social['High school graduate (includes equivalency)']
df_social['college_prop'] = df_social['Some college, no degree'] + df_social["Associate's degree"]
df_social['Year'] = 2015

df_social.head()
```

```python
# reduce columns
social_cols = ['high_school_prop', 'college_prop', 'marriage_prop']
df_social = df_social[['Geography', 'Geographic Area Name', 'Year'] + social_cols]

df_social.head()
```

```python
df_demo = df_demo[['Geography','Geographic Area Name'] + cols]
df_demo
```

```python
# rename and bin columns
df_demo['male_prop'] = df_demo['Male']
df_demo['age_15_24_prop'] = df_demo['15 to 19 years'] + df_demo['20 to 24 years']
df_demo['age_25_44_prop'] = df_demo['25 to 34 years'] + df_demo['35 to 44 years']
df_demo['age_45_64_prop'] = df_demo['45 to 54 years'] + df_demo['55 to 59 years'] + df_demo['60 to 64 years']
df_demo['age_65_prop'] = df_demo['65 to 74 years'] + df_demo['75 to 84 years'] + df_demo['85 years and over']
df_demo['white_prop'] = df_demo['White']
df_demo['black_prop'] = df_demo['Black or African American']
df_demo['asian_prop'] = df_demo['Asian']
df_demo['hispanic_prop'] = df_demo['Hispanic or Latino (of any race)']
df_demo['Year'] = 2015

df_demo
```

```python
# select percent of the following info
# Extract columns where column names contain values in the list
demo_lst = ['Total population']# selected_columns

selected_columns = df_demo2.columns[df_demo2.columns.str.contains('|'.join(demo_lst))]
selected_columns
```

```python
cols = [column for column in selected_columns\
        if 'Estimate!!' in column and "Annotation" not in column]# 
cols
```

```python
# manually select from above list
cols = ['Estimate!!SEX AND AGE!!Total population']

df_demo2 = df_demo2[['Geography','Geographic Area Name'] + cols]
df_demo2
```
