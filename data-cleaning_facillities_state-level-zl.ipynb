```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# display all rows and columns in output
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
data_path = "../02_data/"

df_facility = pd.read_csv(data_path+"treatment_facilities_2016_2020.csv", low_memory=False, dtype='object')
```

# Data cleaning
## number of facilities in each state each year and number of facilities provide a type of service each state each year

```python
df_facility.columns = df_facility.columns.str.lower()
```

```python
df_facility.shape
```

```python
df_facility.head(10)
```

```python
# number of facilities offering substance abuse treatment services
df_tx = df_facility[df_facility['treatmt'] == "1"].groupby(['year', 'state']).size().reset_index(name='n_treatmt')
```

```python
# number of facilities offering naltrexone-related services
df_naltrexone_related = df_facility[(df_facility['srvc71'] == "1") | (df_facility['srvc108'] == "1")].groupby(['year', 'state']).\
             size().reset_index(name='n_naltrexone_related')
```

```python
# number of facilities offering methadone services
df_methadone = df_facility[df_facility['srvc85'] == "1"].groupby(['year', 'state']).\
             size().reset_index(name='n_methadone')
```

```python
# number of facilities offering buprenorphine-related services
df_buprenorphine_related = df_facility[(df_facility['srvc87'] == "1") | (df_facility['srvc86'] == "1")].groupby(['year', 'state']).\
             size().reset_index(name='n_buprenorphine_related')
```

```python
# number of facilities offering opioid treatment program
df_otp = df_facility[df_facility['otp'] == "1"].groupby(['year', 'state']).size().\
reset_index(name='n_otp')
```

```python
# number of facilities offering maintenance services with methadone or buprenorphine.
df_opioidmaint = df_facility[df_facility['opioidmaint'] == "1"].groupby(['year', 'state']).\
             size().reset_index(name='n_opioidmaint')
```

```python
# number of facilities offering maintenancee seervices with medically supervised withdrawal after pre-determined time.
# df_facility[df_facility['opioidwdraw'] == "1"].groupby(['year', 'state']).\
#              size().reset_index(name='n_opioidwdraw')
```

```python
# number of facilities offering computerized substance abuse treatment at least sometimes
df_compsat = df_facility[df_facility['compsat'].isin(["3", "4"])].groupby(['year', 'state']).\
             size().reset_index(name='n_compsat')
df_compsat
```

```python
# number of facilities offering program for DUI/DWI offenders
df_dui = df_facility[df_facility['dui_dwi'] == "1"].groupby(['year', 'state']).\
             size().reset_index(name='n_dui')
```

```python
# number of facilities offering treatment in language other than English
df_multilang = df_facility[df_facility['lang'] == "1"].groupby(['year', 'state']).size().\
reset_index(name='n_multilang')
```

```python
# number of facilities offering flex/lower pay
df_lowpay = df_facility[(df_facility['feescale'] == "1") | (df_facility['payasst'] == "1") | (df_facility['revchk3'] == "1")].groupby(['year', 'state']).\
             size().reset_index(name='n_lowpay')
```

```python
# number of facilities that accept public insurance
df_pubinsur = df_facility[(df_facility['revchk1'] == "1") | (df_facility['revchk8'] == "1")| (df_facility['revchk5'] == "1")|\
           (df_facility['revchk10'] == "1") | (df_facility['revchk15'] == "1")].groupby(['year', 'state']).\
             size().reset_index(name='n_pubinsur')
```

```python
# number of facilities that are licensed or accredited
df_licen = df_facility[(df_facility['licen'] == "1") | (df_facility['accred'] == "1")].groupby(['year', 'state']).\
             size().reset_index(name='n_licen')
```

```python
df_fac = df_tx.merge(df_naltrexone_related, on = ["year", "state"], how="outer").\
merge(df_methadone, on = ["year", "state"], how="outer").\
merge(df_buprenorphine_related, on = ["year", "state"], how="outer").\
merge(df_otp, on = ["year", "state"], how="outer").\
merge(df_opioidmaint, on = ["year", "state"], how="outer").\
merge(df_compsat, on = ["year", "state"], how="outer").\
merge(df_dui, on = ["year", "state"], how="outer").\
merge(df_multilang, on = ["year", "state"], how="outer").\
merge(df_lowpay, on = ["year", "state"], how="outer").\
merge(df_pubinsur, on = ["year", "state"], how="outer").\
merge(df_licen, on = ["year", "state"], how="outer")
```

```python
(df_fac.groupby(["state", "year"]).count()>1).any().any()
```

```python
df_fac_2016 = df_fac[df_fac['year']=="2016"]
df_fac_2017 = df_fac[df_fac['year']=="2017"]
df_fac_2018 = df_fac[df_fac['year']=="2018"]
df_fac_2019 = df_fac[df_fac['year']=="2019"]
df_fac_2020 = df_fac[df_fac['year']=="2020"]
```

```python
writer = pd.ExcelWriter('N_facillities_by_state_year.xlsx', engine='xlsxwriter')

df_fac_2016.to_excel(writer, sheet_name='2016', index=False)
df_fac_2017.to_excel(writer, sheet_name='2017', index=False)
df_fac_2018.to_excel(writer, sheet_name='2018', index=False)
df_fac_2019.to_excel(writer, sheet_name='2019', index=False)
df_fac_2020.to_excel(writer, sheet_name='2020', index=False)
writer.save()
```
